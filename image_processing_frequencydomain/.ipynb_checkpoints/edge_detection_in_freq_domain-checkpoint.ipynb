```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import color, img_as_float, filters
from skimage.io import imread
img = img_as_float(imread("photo.jpeg", as_gray=True))
gray = color.rgb2gray(img) if img.ndim == 3 else img
gray = img_as_float(gray)

# 2) High-pass (frequency domain)
edges_hp = filters.butterworth(
    gray,
    cutoff_frequency_ratio=0.03,  #  daha agresif (daha ok edge + grlt)
    high_pass=True,
    order=2
)

# 3) Grsel olarak daha anlalr olmas iin normalize (opsiyonel)
edges_hp_n = (edges_hp - edges_hp.min()) / (edges_hp.max() - edges_hp.min() + 1e-9)

# 4) Grselletir
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.imshow(gray, cmap="gray")
plt.title("Gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(edges_hp_n, cmap="gray")
plt.title("Edges (Butterworth High-Pass)")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# normalize yardmc fonksiyonu
def normalize(x):
    return (x - x.min()) / (x.max() - x.min() + 1e-9)

edges_hp1 = filters.butterworth(gray, cutoff_frequency_ratio=0.02, high_pass=True, order=2)
edges_hp2 = filters.butterworth(gray, cutoff_frequency_ratio=0.05, high_pass=True, order=2)
edges_hp3 = filters.butterworth(gray, cutoff_frequency_ratio=0.03, high_pass=True, order=4)

edges_hp1 = normalize(edges_hp1)
edges_hp2 = normalize(edges_hp2)
edges_hp3 = normalize(edges_hp3)

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(edges_hp1, cmap="gray")
plt.title("HPF: cutoff=0.02, order=2")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(edges_hp2, cmap="gray")
plt.title("HPF: cutoff=0.05, order=2")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(edges_hp3, cmap="gray")
plt.title("HPF: cutoff=0.03, order=4")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def normalize(x):
    return (x - x.min()) / (x.max() - x.min() + 1e-9)

# normalize
e1 = normalize(edges_hp1)
e2 = normalize(edges_hp2)
e3 = normalize(edges_hp3)

# ===============================
# SADECE BURAYI DETR
# ===============================
EDGE_THRESHOLD = 0.25   # 0.15  0.40 aras dene

# binary edges
b1 = e1 > EDGE_THRESHOLD
b2 = e2 > EDGE_THRESHOLD
b3 = e3 > EDGE_THRESHOLD

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(b1, cmap="gray")
plt.title("Binary Edge (cutoff=0.02, order=2)")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(b2, cmap="gray")
plt.title("Binary Edge (cutoff=0.05, order=2)")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(b3, cmap="gray")
plt.title("Binary Edge (cutoff=0.03, order=4)")
plt.axis("off")

plt.tight_layout()
plt.show()
```
