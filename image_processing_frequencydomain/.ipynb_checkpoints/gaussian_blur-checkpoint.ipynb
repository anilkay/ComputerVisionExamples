```python
from skimage import filters, img_as_float
from skimage.io import imread,imshow

img = img_as_float(imread("photo.jpeg"))

img_low = filters.gaussian(
    img,
    sigma=2,
    channel_axis=-1
)
img_low_more_blur=img_low = filters.gaussian(
    img,
    sigma=6,
    channel_axis=-1
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title("Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(img_low)
plt.title("Gaussian Blur (sigma=2)")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(img_low_more_blur)
plt.title("Gaussian Blur (sigma=6)")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from skimage import color, filters

gray = color.rgb2gray(img)

thresh = filters.threshold_otsu(gray)
binary = gray > thresh
from skimage.measure import label, regionprops

labels = label(binary)
regions = regionprops(labels)

# En byk alanl blge
largest_region = max(regions, key=lambda r: r.area)

foreground_mask = labels == largest_region.label
from skimage.morphology import binary_closing, disk

foreground_mask = binary_closing(foreground_mask, disk(5))
from skimage import filters

blurred = filters.gaussian(img, sigma=6, channel_axis=-1)
result = img.copy()
result[~foreground_mask] = blurred[~foreground_mask]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(foreground_mask, cmap="gray")
plt.title("Foreground Mask (Biggest Object)")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(result)
plt.title("Background Blurred")
plt.axis("off")

plt.tight_layout()
plt.show()
```
