```python
from skimage.io import imread
from skimage import img_as_float
from skimage.restoration import wiener
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# load grayscale image
img = img_as_float(imread("photo.jpeg", as_gray=True))

# PSF: blur kernel (5x5)
psf = np.ones((5,5)) / 25  

# create blurred version
blurred = convolve2d(img, psf, mode='same', boundary='symm')

# add noise
noisy = blurred + 0.01 * np.random.randn(*blurred.shape)

# apply Wiener filter
restored_img = wiener(noisy, psf, balance=0.1)

# plot
fig, axes = plt.subplots(1,3, figsize=(12,4))
axes[0].imshow(img, cmap='gray'); axes[0].set_title("Original"); axes[0].axis('off')
axes[1].imshow(noisy, cmap='gray'); axes[1].set_title("Blur+Noise"); axes[1].axis('off')
axes[2].imshow(restored_img.real, cmap='gray'); axes[2].set_title("Wiener Restored"); axes[2].axis('off')
plt.show()
```

```python
# It works minimize the random blur. 
```

```python
from skimage.io import imread
from skimage import img_as_float
from skimage.restoration import wiener
from skimage.filters import gaussian, median
from skimage.morphology import square

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# 1) GRNTY YKLE
img = img_as_float(imread("photo.jpeg", as_gray=True))

# 2) BLUR + NOISE OLUTUR
psf = np.ones((5,5)) / 25
blurred = convolve2d(img, psf, mode='same', boundary='symm')
noisy = blurred + 0.01 * np.random.randn(*blurred.shape)

# 3) WIENER
rest_wiener = wiener(noisy, psf, balance=0.1)

# 4) LINEAR: GAUSSIAN FILTER
rest_gauss = gaussian(noisy, sigma=1)

# 5) NONLINEAR: MEDIAN FILTER
rest_median = median(noisy, square(3))

# 6) GSTER
fig, axes = plt.subplots(1,5, figsize=(18,4))
axes[0].imshow(img, cmap='gray');        axes[0].set_title("Original");      axes[0].axis('off')
axes[1].imshow(noisy, cmap='gray');      axes[1].set_title("Blur+Noise");    axes[1].axis('off')
axes[2].imshow(rest_wiener.real, cmap='gray'); axes[2].set_title("Wiener");  axes[2].axis('off')
axes[3].imshow(rest_gauss, cmap='gray'); axes[3].set_title("Gaussian");      axes[3].axis('off')
axes[4].imshow(rest_median, cmap='gray');axes[4].set_title("Median");        axes[4].axis('off')
plt.show()
```

```python
# Wener has the best performance.
```

```python
from skimage.io import imread
from skimage import img_as_float
from skimage.restoration import wiener
from skimage.morphology import square
from skimage.util import random_noise
from skimage.filters import median

import numpy as np
import matplotlib.pyplot as plt

# 1) ORJNAL YKLE
img = img_as_float(imread("photo.jpeg", as_gray=True))

# 2) SALTPEPPER GRLT EKLE
sp_noisy = random_noise(img, mode="s&p", amount=0.1)  # amount: grlt oran

# 3) MEDIAN FLTRE (NONLINEAR)
med_restored = median(sp_noisy, square(3))

# 4) WIENER FLTRE (GAUSS VARSAYIMINA RAMEN DENEYSEL)
# Basit bir PSF: 3x3 average kernel
psf = np.ones((3, 3)) / 9
wiener_restored = wiener(sp_noisy, psf, balance=0.1)

# 5) GSTER
fig, axes = plt.subplots(1, 4, figsize=(18, 4))
axes[0].imshow(img, cmap='gray');              axes[0].set_title("Original"); axes[0].axis('off')
axes[1].imshow(sp_noisy, cmap='gray');         axes[1].set_title("Salt & Pepper"); axes[1].axis('off')
axes[2].imshow(med_restored, cmap='gray');     axes[2].set_title("Median"); axes[2].axis('off')
axes[3].imshow(wiener_restored.real, cmap='gray'); axes[3].set_title("Wiener"); axes[3].axis('off')
plt.show()
```

```python
# In Salt and pepper noise median is better. Wiener filter does nothing.
```

```python
from skimage import data, img_as_float
from skimage.util import random_noise
from skimage.filters import median
from skimage.morphology import square
import matplotlib.pyplot as plt
from scipy.ndimage import uniform_filter


def lee_filter(img, size=5):
    img = img.astype(np.float32)

    # local mean
    mean_local = uniform_filter(img, size)
    # local mean of squared
    mean_sq_local = uniform_filter(img**2, size)
    # local variance
    var_local = mean_sq_local - mean_local**2

    # noise variance (ok basit tahmin)
    noise_var = np.mean(var_local)

    # Lee gain
    k = var_local / (var_local + noise_var + 1e-12)
    out = mean_local + k * (img - mean_local)
    return out

img = img_as_float(imread("photo.jpeg", as_gray=True))

# add speckle noise (multiplicative)
speckle = random_noise(img, mode="speckle", var=0.1)

# apply median filter
rest_median = median(speckle, square(3))
lee_restored = lee_filter(speckle, size=5)

# show
fig, axes = plt.subplots(1,4, figsize=(12,4))
axes[0].imshow(img, cmap='gray'); axes[0].set_title("Original"); axes[0].axis('off')
axes[1].imshow(speckle, cmap='gray'); axes[1].set_title("Speckle Noise"); axes[1].axis('off')
axes[2].imshow(rest_median, cmap='gray'); axes[2].set_title("Median Restored"); axes[2].axis('off')
axes[3].imshow(lee_restored, cmap='gray'); axes[3].set_title("Median"); axes[3].axis('off')

plt.show()
```
